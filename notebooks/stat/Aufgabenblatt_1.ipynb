 1. Geben Sie drei Moeglichkeiten an, den Vektor (1, 4, 7, 1, 4, 7) zu erzeugen.

```python
import numpy as np

# Moeglichkeit 1: Direkt angeben
vektor_1 = np.array([1, 4, 7, 1, 4, 7])

# Moeglichkeit 2: Durch wiederholtes Append
vektor_2 = np.concatenate(([1, 4, 7], [1, 4, 7]))

# Moeglichkeit 3: Mit tile-Funktion zur Wiederholung eines Musters
vektor_3 = np.tile([1, 4, 7], 2)

vektor_1, vektor_2, vektor_3
```

2. Erzeugen Sie einen Vektor namens Groesse mit den Werten (S, M, L) und einen
weiteren Vektor namens Anzahl mit den Werten (130,210,170). <br>
a) Zeichnen Sie zwei zugehoerige Histogramme, einmal mit den aboluten und mit den relativen Haeufigkeiten.<br>
b) Ersetzen Sie die Labels (S, M, L) durch (34, 38, 42) und machen Sie das gleiche wie in a). <br>
c) Berechnen Sie drei statistische Kennwerte mithilfe Python Funktionen. <br>
d) FÃ¼gen Sie die beiden Variablen Groesse und Anzahl zu einem Data Frame namens Daten
zusammen.

```python
import matplotlib.pyplot as plt
import pandas as pd

# a) Erzeugen der Vektoren
groesse = ['S', 'M', 'L']
anzahl = [130, 210, 170]

# Zeichnen der Histogramme (absolute Haeufigkeiten)
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.bar(groesse, anzahl)
plt.title("Absolut Haeufigkeiten")
plt.xlabel("Groesse")
plt.ylabel("Anzahl")

# Relativ Haeufigkeiten berechnen und zeichnen
relative_haeufigkeiten = np.array(anzahl) / sum(anzahl)
plt.subplot(1, 2, 2)
plt.bar(groesse, relative_haeufigkeiten)
plt.title("Relative Haeufigkeiten")
plt.xlabel("Groesse")
plt.ylabel("Relative Haeufigkeit")

plt.tight_layout()
plt.show()

# b) Ersetzen der Labels (S, M, L) durch (34, 38, 42)
groesse_numeric = [34, 38, 42]

# Histogramme mit numerischen Groessen
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.bar(groesse_numeric, anzahl)
plt.title("Absolut Haeufigkeiten (numerisch)")
plt.xlabel("Groesse (Numerisch)")
plt.ylabel("Anzahl")

plt.subplot(1, 2, 2)
plt.bar(groesse_numeric, relative_haeufigkeiten)
plt.title("Relative Haeufigkeiten (numerisch)")
plt.xlabel("Groesse (Numerisch)")
plt.ylabel("Relative Haeufigkeit")

plt.tight_layout()
plt.show()

# c) Berechnung von Mittelwert, Median und Standardabweichung
mean = np.mean(anzahl)
median = np.median(anzahl)
std_dev = np.std(anzahl)

mean, median, std_dev
```

4. Simulieren Sie eine genÃ¼gend grosse Ziehung N aus den Ziffern 0 bis 9. <br>
a) Stellen Sie das Ergebnis in einem passenden Format graphisch dar. <br>
b) Berechnen Sie mithilfe einer Simulation die Wahrscheinlichkeit eine Ziffer grÃ¶sser als 6 zu ziehen. <br>
c) Das Gesetz der grossen Zahlen besagt, dass das arithemtische Mittel der Stichprobe mit grossem N zum Mittelwert, welches hier 4.5 betraegt, konvergiert. ÃberprÃ¼fen Sie das mit Ihrer Stichprobe. <br>

```python
import numpy as np

# Seed fÃ¼r Reproduzierbarkeit
np.random.seed(42)

# Simulationsparameter
N = 10000  # GroÃe Anzahl an Ziehungen
ziehungen = np.random.randint(0, 10, N)

# a) Grafische Darstellung der Ziehungen (HÃ¤ufigkeiten der Ziffern)
plt.figure(figsize=(8, 6))
plt.hist(ziehungen, bins=np.arange(11)-0.5, edgecolor='black', density=True)
plt.xticks(np.arange(10))
plt.title('HÃ¤ufigkeiten der Ziffern 0 bis 9 bei N Ziehungen')
plt.xlabel('Ziffer')
plt.ylabel('Relative HÃ¤ufigkeit')
plt.show()

# b) Wahrscheinlichkeit, eine Ziffer grÃ¶Ãer als 6 zu ziehen
prob_greater_6 = np.sum(ziehungen > 6) / N

# c) ÃberprÃ¼fung des Gesetzes der groÃen Zahlen (Konvergenz des Mittelwerts)
mean_sample = np.mean(ziehungen)
expected_mean = 4.5  # Der theoretische Mittelwert

prob_greater_6, mean_sample, expected_mean
```
