### Aufgabe 1

Berechnen Sie den Inhalt folgender beschränkter Menge des
$\mathbb{R}^3$:

$$
M=\{(x_1,x_2,x_3)|0 \leq x_1 \leq 1, 0 \leq x_2 \leq x_1,0
\leq x_3 \leq x_1x_2\}
$$

$$\int_{0}^{1} \int_{0}^{x_1} \int_{0}^{x_1 x_2} 1 \, dx_3 \, dx_2 \, dx_1$$

$$\int_{0}^{x_1 x_2} 1 \, dx_3 = x_1 x_2$$

$$\int_{0}^{x_1} x_1 x_2 \, dx_2 = x_1 \cdot \left[ \frac{x_2^2}{2} \right]_{0}^{x_1}$$
$$= x_1 \cdot \frac{x_1^2}{2} = \frac{x_1^3}{2}$$

$$\int_{0}^{1} \frac{x_1^3}{2} \, dx_1 = \frac{1}{2} \cdot \int_{0}^{1} x_1^3 \, dx_1$$
$$= \frac{1}{2} \cdot \left[ \frac{x_1^4}{4} \right]_{0}^{1}$$
$$= \frac{1}{2} \cdot \frac{1}{4} = \frac{1}{8}$$

```python
x1, x2, x3 = symbols('x1 x2 x3')
f = 1
limits = [(x3, 0, x1*x2), (x2, 0, x1), (x1, 0, 1)]

result = integrate(f, *limits)
result
```

### Aufgabe 2

Berechnen Sie den Inhalt folgender beschränkter Menge des
$\mathbb{R}^3$:

$$
M=\{(x_1,x_2,x_3)|0 \leq x_1 \leq 1, 0 \leq x_2 \leq 1,0
\leq x_3 \leq \min{(x_1,x_2)}\}
$$

$$\int_{0}^{1} \int_{0}^{1} \int_{0}^{\min(x_1, x_2)} 1 \, dx_3 \, dx_2 \, dx_1$$
$$\int_{0}^{\min(x_1, x_2)} 1 \, dx_3 = \min(x_1, x_2) - 0 = \min(x_1, x_2)$$
$$\int_{0}^{1} \int_{0}^{1} \min(x_1, x_2) \, dx_2 \, dx_1$$
$$\int_{0}^{1} \int_{0}^{1} \min(x_1, x_2) \, dx_2 \, dx_1 = \int_{0}^{1} \left[ x_2 \min(x_1, x_2) \right]_{0}^{1} \, dx_1$$
$$\int_{0}^{1} \left[ x_2 \min(x_1, x_2) \right]_{0}^{1} \, dx_1 = \int_{0}^{1} x_1 \, dx_1 = \frac{1}{2}$$

```python
f = 1
limits = [(x3, 0, Min(x1, x2)), (x2, 0, 1), (x1, 0, 1)]

result = integrate(f, *limits)
result
```

### Aufgabe 3

Berechnen Sie den Inhalt folgender beschränkter Menge des
$\mathbb{R}^3$:

$$
M=\{(x_1,x_2,x_3)|0 \leq x_1 \leq h, |x_2| \leq ax_1,|x_3| \leq ax_1\};a,h\in \mathbb{R},a>0,h>0
$$

$$\int_{0}^{h} \int_{-a x_1}^{a x_1} \int_{-a x_1}^{a x_1} 1 \, dx_3 \, dx_2 \, dx_1$$
$$\int_{0}^{h} \int_{-a x_1}^{a x_1} \int_{-a x_1}^{a x_1} 1 \, dx_3 \, dx_2 \, dx_1$$
$$\int_{-a x_1}^{a x_1} 1 \, dx_3 = 2 a x_1$$
$$\int_{-a x_1}^{a x_1} 2 a x_1 \, dx_2 = 4 a^2 x_1^2$$
$$\int_{0}^{h} \int_{-a x_1}^{a x_1} 4 a^2 x_1^2 \, dx_2 \, dx_1$$
$$\int_{0}^{h} 4 a^2 x_1^2 \cdot 2 a x_1 \, dx_1 = 8 a^3 \int_{0}^{h} x_1^3 \, dx_1$$
$$8 a^3 \cdot \left[ \frac{x_1^4}{4} \right]_{0}^{h} = 8 a^3 \cdot \frac{h^4}{4} = 2 a^3 h^4$$

```python
x1, x2, x3, a, h = symbols('x1 x2 x3 a h')
f = 1
limits = [(x2, -a*x1, a*x1), (x3, -a*x1, a*x1), (x1, 0, h)]

result = integrate(f, *limits)
result
```

### Aufgabe 4

Berechnen Sie $\int_Bf(x)dx$ mit $x=(x_1,\ldots,x_n) \in
\mathbb{R}^n$ mit folgender Angabe:

$$
f(x_1,x_2)=x_1+x_2, B=\{0 \leq x_1 \leq 1, 0 \leq x_2 \leq x_1\}
$$

$$\int_{0}^{1} \int_{0}^{x_1} (x_1 + x_2) \, dx_2 \, dx_1$$

$$\int_{0}^{x_1} (x_1 + x_2) \, dx_2 = \left[ x_1 x_2 + \frac{x_2^2}{2} \right]_{0}^{x_1}$$
$$= \left( x_1 \cdot x_1 + \frac{x_1^2}{2} \right) - \left( 0 + 0 \right)$$
$$= x_1^2 + \frac{x_1^2}{2}$$
$$= \frac{3}{2} x_1^2$$

$$\int_{0}^{1} \frac{3}{2} x_1^2 \, dx_1$$

$$\int_{0}^{1} \frac{3}{2} x_1^2 \, dx_1 = \frac{3}{2} \cdot \left[ \frac{x_1^3}{3} \right]_{0}^{1}$$
$$= \frac{3}{2} \cdot \frac{1}{3}$$
$$= \frac{1}{2}$$

```python
x1, x2 = symbols('x1 x2')
f = x1 + x2
limits = [(x2, 0, x1), (x1, 0, 1)]

result = integrate(f, *limits)
result
```

### Aufgabe 5

Berechnen Sie $\int_Bf(x)dx$ mit $x=(x_1,\ldots,x_n) \in
\mathbb{R}^n$ mit folgender Angabe:

$$
f(x_1,x_2)=\frac{1}{x_1}+\frac{1}{x_2}, B=\{1 \leq x_1 \leq 2, 1 \leq x_2 \leq 2\}
$$

Plotten Sie $\int_Bf(x)dx$.

$$\int_{1}^{2} \int_{1}^{2} \left( \frac{1}{x_1} + \frac{1}{x_2} \right) \, dx_2 \, dx_1$$

$$\int_{1}^{2} \left( \frac{1}{x_1} + \frac{1}{x_2} \right) \, dx_2 = \left[ \frac{1}{x_1} \cdot x_2 + \ln(x_2) \right]_{1}^{2}$$
$$= \frac{1}{x_1} \cdot 2 + \ln(2) - \left( \frac{1}{x_1} \cdot 1 + \ln(1) \right)$$
$$= \frac{1}{x_1} + \ln(2)$$

$$\int_{1}^{2} \left( \frac{1}{x_1} + \ln(2) \right) \, dx_1 = \left[ \ln(x_1) + x_1\ln(2) \right]_{1}^{2}$$
$$= (\ln(2) + 2\ln(2)) - (\ln(1) + \ln(2))$$
$$= 2\ln(2) = \ln(4)$$

```python
f = 1/x1 + 1/x2
limits = [(x1, 1, 2), (x2, 1, 2)]

result = integrate(f, *limits)

print(result)

p = plot3d(f, (x1, 1, 2), (x2, 1, 2), show=False)
p.title = 'Plot von f(x1, x2)'
p.show()
```

### Aufgabe 6

Berechnen Sie $\int_Bf(x)dx$ mit $x=(x_1,\ldots,x_n) \in
\mathbb{R}^n$ mit folgender Angabe:

$$
f(x_1,x_2,x_3)=x_1x_2x_3, B=\{0 \leq x_1 \leq 1, x_2 \geq 0, x_3
\geq 0, 0 \leq x_2+x_3\leq 1\}$$


Plotten Sie $B$.

$$\int_{0}^{1} \int_{0}^{1-x_3} \int_{0}^{1} x_1 x_2 x_3 \, dx_1 \, dx_2 \, dx_3$$

$$\int_{0}^{1} x_1 x_2 x_3 \, dx_1 = x_2 x_3 \cdot \left[ \frac{x_1^2}{2} \right]_{0}^{1}$$
$$= x_2 x_3 \cdot \frac{(1)^2}{2}$$

$$\int_{0}^{1-x_3} x_2 x_3 \cdot \frac{1}{2} \, dx_2 = x_3 \cdot \left[ \frac{x_2^3}{6} \right]_{0}^{1-x_3}$$
$$= x_3 \cdot \frac{(1-x_3)^3}{6}$$

$$\int_{0}^{1} x_3 \cdot \frac{(1-x_3)^3}{6} \, dx_3 = \frac{1}{6} \cdot \int_{0}^{1} x_3 (1-x_3)^3 \, dx_3$$

```python
f = x1 * x2 * x3
limits = [(x1, 0, 1), (x2, 0, 1-x3), (x3, 0, 1)]

result = integrate(f, *limits)
result
```
