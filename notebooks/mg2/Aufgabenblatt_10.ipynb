### Aufgabe 1

Bestimmen Sie die Definitheit der folgenden quadratischen Form
$\boldsymbol{x}^tA \boldsymbol{x}$:

$$
A= \left( \begin{array}{ccc}
                                      1 & 2 & 1\\
                                      2 & 1 & 1\\
                                      1 & 1 & 4\\
                                    \end{array} \right)
$$


$$
\text{Eigenwerte: } \lambda
$$
$$
\det(A - \lambda I) = \begin{vmatrix}
1 - \lambda & 2 & 1 \\
2 & 1 - \lambda & 1 \\
1 & 1 & 4 - \lambda
\end{vmatrix} = 0
$$
$$
\begin{align}
\begin{vmatrix}
1 - \lambda & 2 & 1 \\
2 & 1 - \lambda & 1 \\
1 & 1 & 4 - \lambda
\end{vmatrix} &= (1 - \lambda) \left[(1 - \lambda)(4 - \lambda) - 1 \right] - 2 \left[2(4 - \lambda) - 1 \right] + 1 \left[2 - 2(1 - \lambda) \right] \\
&= (1 - \lambda) \left[4 - \lambda - 4\lambda + \lambda^2 - 1 \right] - 2 \left[8 - 2\lambda - 1 \right] + 1 \left[2 - 2 + 2\lambda \right] \\
&= (1 - \lambda)(\lambda^2 - 5\lambda + 3) - 2(7 - 2\lambda) + 2\lambda \\
&= (1 - \lambda)(\lambda^2 - 5\lambda + 3) - 14 + 4\lambda + 2\lambda \\
&= \lambda^3 - 5\lambda^2 + 3\lambda - \lambda^2 + 5\lambda - 3 - 14 + 6\lambda \\
&= \lambda^3 - 6\lambda^2 + 14\lambda - 17 = 0
\end{align}
$$
$$
\text{Lösungen der charakteristischen Gleichung: } \lambda_1 = 5, \lambda_2 = -2, \lambda_3 = 2
$$
$$
\text{Eigenwerte: } \lambda_1 = 5, \lambda_2 = -2, \lambda_3 = 2
$$
$$
\text{Definitheit: } \text{\textbf{indefinit} (da sowohl positive als auch negative Eigenwerte)}
$$

### Aufgabe 4

Bestimmen Sie die lokalen Extrema von $\boldsymbol{x}^tA \boldsymbol{x}+\boldsymbol{b}^t \boldsymbol{x}, A= \left(
\begin{array}{cc}
                                      1 & \frac{1}{2}\\
                                      \frac{1}{2} & 1\\
                                    \end{array} \right), \boldsymbol{b} = \left( \begin{array}{c}
                                      -3\\
                                      -2\\
                                    \end{array} \right)$ und geben Sie mit Hilfe der Hesse'schen Matrix an, ob Minima oder
Maxima vorliegen.


$$
f(\boldsymbol{x}) = \boldsymbol{x}^t A \boldsymbol{x} + \boldsymbol{b}^t \boldsymbol{x}
$$
$$
\text{Gradient berechnen:}
\nabla f = \begin{pmatrix}
\frac{\partial f}{\partial x_1} \\
\frac{\partial f}{\partial x_2}
\end{pmatrix} = A \boldsymbol{x} + \boldsymbol{b}
$$
$$
\nabla f = 0 \Rightarrow A \boldsymbol{x} + \boldsymbol{b} = 0 \Rightarrow \boldsymbol{x} = -A^{-1} \boldsymbol{b}
$$
$$
\begin{align}
A^{-1} &= \frac{1}{\det(A)} \text{adj}(A) \\
&= \frac{1}{1 \cdot 1 - \left(\frac{1}{2} \cdot \frac{1}{2}\right)} \begin{pmatrix}
1 & -\frac{1}{2} \\
-\frac{1}{2} & 1
\end{pmatrix} \\
&= \frac{1}{\frac{3}{4}} \begin{pmatrix}
1 & -\frac{1}{2} \\
-\frac{1}{2} & 1
\end{pmatrix} \\
&= \frac{4}{3} \begin{pmatrix}
1 & -\frac{1}{2} \\
-\frac{1}{2} & 1
\end{pmatrix} \\
&= \begin{pmatrix}
\frac{4}{3} & -\frac{2}{3} \\
-\frac{2}{3} & \frac{4}{3}
\end{pmatrix}
\end{align}
$$
$$
\text{Kritischer Punkt:}
\boldsymbol{x} = -\begin{pmatrix}
\frac{4}{3} & -\frac{2}{3} \\
-\frac{2}{3} & \frac{4}{3}
\end{pmatrix} \begin{pmatrix}
-3 \\
-2
\end{pmatrix}
= \begin{pmatrix}
\frac{4}{3} \cdot 3 + \frac{2}{3} \cdot 2 \\
\frac{2}{3} \cdot 3 + \frac{4}{3} \cdot 2
\end{pmatrix}
= \begin{pmatrix}
4 \\
2
\end{pmatrix}
$$
$$
\text{Hesse-Matrix:}
H_f = A
$$
$$
\text{Da die Eigenwerte von } A \text{ positiv sind (} \lambda_1 = \frac{3}{2}, \lambda_2 = \frac{1}{2} \text{), ist } A \text{ positiv definit.}
$$
$$
\text{Das bedeutet, der kritische Punkt } \boldsymbol{x} = \begin{pmatrix} 4 \\ 2 \end{pmatrix} \text{ ist ein Minimum.}
$$

```python
x1, x2 = sp.symbols('x1 x2')
A4 = sp.Matrix([
    [1, 1/2],
    [1/2, 1]
])
b4 = sp.Matrix([-3, -2])

# Quadratische Form und lineare Form
f4 = sp.Matrix([x1, x2]).T * A4 * sp.Matrix([x1, x2]) + b4.T * sp.Matrix([x1, x2])
f4 = f4[0]

# Berechnung der kritischen Punkte
grad_f4 = [sp.diff(f4, var) for var in (x1, x2)]
critical_points = sp.solve(grad_f4, (x1, x2))
critical_points
```

### Aufgabe 6

Untersuchen Sie folgende Funktionen auf Extrema und Sattelpunkte:

a) $f(x,y)=x^2+y^2, D_f=\mathbb{R}^2$  
b) $f(x,y)=x^2-y^2, D_f=\mathbb{R}^2$

Plotten Sie die Funktionen.

## a)

$$
f(x,y) = x^2 + y^2
$$
$$
\text{Gradient berechnen:}
\nabla f = \begin{pmatrix}
\frac{\partial f}{\partial x} \\
\frac{\partial f}{\partial y}
\end{pmatrix} = \begin{pmatrix}
2x \\
2y
\end{pmatrix}
$$
$$
\nabla f = 0 \Rightarrow 2x = 0, \quad 2y = 0 \Rightarrow (x, y) = (0, 0)
$$
$$
\text{Hesse-Matrix:}
H_f = \begin{pmatrix}
\frac{\partial^2 f}{\partial x^2} & \frac{\partial^2 f}{\partial x \partial y} \\
\frac{\partial^2 f}{\partial y \partial x} & \frac{\partial^2 f}{\partial y^2}
\end{pmatrix} = \begin{pmatrix}
2 & 0 \\
0 & 2
\end{pmatrix}
$$
$$
\text{Eigenwerte der Hesse-Matrix:}
\lambda_1 = 2, \lambda_2 = 2
$$
$$
\text{Da die Eigenwerte positiv sind, ist die Hesse-Matrix positiv definit.}
$$
$$
\text{Das bedeutet, der kritische Punkt } (0,0) \text{ ist ein Minimum.}
$$

```python
x, y = sp.symbols('x y')
f6a = x**2 + y**2

# Berechnung der kritischen Punkte
grad_f6a = [sp.diff(f6a, var) for var in (x, y)]
critical_points_6a = sp.solve(grad_f6a, (x, y))
critical_points_6a

# Plotten der Funktion
sp.plotting.plot3d(f6a, (x, -5, 5), (y, -5, 5))
```

## b)

$$
f(x,y) = x^2 - y^2
$$
$$
\text{Gradient berechnen:}
\nabla f = \begin{pmatrix}
\frac{\partial f}{\partial x} \\
\frac{\partial f}{\partial y}
\end{pmatrix} = \begin{pmatrix}
2x \\
-2y
\end{pmatrix}
$$
$$
\nabla f = 0 \Rightarrow 2x = 0, \quad -2y = 0 \Rightarrow (x, y) = (0, 0)
$$
$$
\text{Hesse-Matrix:}
H_f = \begin{pmatrix}
\frac{\partial^2 f}{\partial x^2} & \frac{\partial^2 f}{\partial x \partial y} \\
\frac{\partial^2 f}{\partial y \partial x} & \frac{\partial^2 f}{\partial y^2}
\end{pmatrix} = \begin{pmatrix}
2 & 0 \\
0 & -2
\end{pmatrix}
$$
$$
\text{Eigenwerte der Hesse-Matrix:}
\lambda_1 = 2, \lambda_2 = -2
$$
$$
\text{Da die Eigenwerte sowohl positiv als auch negativ sind, ist die Hesse-Matrix indefinit.}
$$
$$
\text{Das bedeutet, der kritische Punkt } (0,0) \text{ ist ein Sattelpunkt.}
$$

```python
x, y = sp.symbols('x y')
f6a = x**2 - y**2

# Berechnung der kritischen Punkte
grad_f6a = [sp.diff(f6a, var) for var in (x, y)]
critical_points_6a = sp.solve(grad_f6a, (x, y))
critical_points_6a

# Plotten der Funktion
sp.plotting.plot3d(f6a, (x, -5, 5), (y, -5, 5))
```
