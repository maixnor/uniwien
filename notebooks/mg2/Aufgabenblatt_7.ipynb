## Eindimensionale Analysis – Bestimmte Integrale

```python
from sympy import *
x = symbols("x", real = true)
```

```python
def plot_functions(f1, f2, x_min, x_max, y_min, y_max):
    p = plot(f1, xlim=(x_min, x_max), ylim=(y_min, y_max), show=False, label="f", legend=true)
    p.extend(plot(f2, xlim=(x_min, x_max), ylim=(y_min, y_max), show=False, label="g"))
    p.show()
```

### Aufgabe 1

Berechnen Sie die Fläche zwischen den Funktionen $f$ und $g$ im
Intervall zwischen den reellwertigen Schnittpunkten beider
Funktionen:

$$
f(x)=x^2,g(x)=\sqrt{x}
$$

Plotten Sie die beiden Funktionen.

```python
f = x**2
g = sqrt(x)

plot_functions(f, g, -1, 2, -1, 2)
```

nachdem wir einen Blick auf das Plot geworfen haben, sehen wir die Schnittpunkte. Die kann man gleich berechnen.

$$f = g \rightarrow x^2 = \sqrt{x}$$
$$0 = x^2 - \sqrt{x} = x^4 - x$$
$$x_1 = 0, x_2 = 1$$

damit haben wir unser bestimmtes Integral von 0 bis 1.

Zu berechnen ist die Flaeche unter g - Flaeche unter f, da f unter g liegt.

$$\int_0^1 g - f dx = \int_0^1 \sqrt{x} - x^2 dx$$
$$\frac{2}{3} - \frac{1}{3} = \frac{1}{3}$$

```python
fi = integrate(f, (x, 0, 1))
gi = integrate(g, (x, 0, 1))

print(fi)
print(gi)

gi - fi
```

### Aufgabe 7

Dreht man eine Kurve mit $y=f(x)$ um die $x$-Achse, so ist die
Mantelfläche des entstehenden Drehkörpers:

$$M=2\pi\int_a^b{y\sqrt{1+y'^2}dx}$$

Eine kubische Parabel mit $y=x^3$ rotiert um die $x$-Achse. Wie groß
ist die Mantelfläche für $a=0$ und $b=1$?

Plotten Sie die Parabel.

```python
f = x**3

plot(f, xlim=(-5, 5), ylim=(0, 10), label="f", legend=true)
```

```python
fd = diff(f, x)
fd2 = fd * fd
g = f * sqrt(1 + fd2)

M = 2 * pi * integrate(g, (x, 0, 1))
M
```
